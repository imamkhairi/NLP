### jupyter notebook の使い方

jupyter notebook：OSSのWebアプリケーション．コードを埋め込んだり実行したりでき，また，ドキュメントを作成することができる．(OSS:Open Source Software)<br>
本授業では，プログラミング言語pythonのコード作成・実行する，また，マークダウン言語(HTML形式)やLaTeXで数式，図の挿入などをメモをすることに使う．<br>
pythonコードはnotebookで作成した場合， ファイルの拡張子は，（.ipynb) となります．<br>

コマンドラインやエディタを使ってpythonを実行する場合は，(.py)の拡張子を付けます．同じコードで実行できます．ただし，利用するパッケージやモジュールがインストールされていることが必要です．<br>

#### 4. jupyter notebook を終了する方法
開いているpythonコードのタブの✖︎をクリックし閉じ， notebookを起動したときのフォルダやファイルが見えているタブを選択，
- (1) jupyter タブの [Running] を選択
- (2) 開いた各ファイルのshutdown タブを選択し，　ファイルをひとつひとつクローズする．
- (3) 右上 [Logout] タブを選択．
- (4) 端末に戻り，　Ctr-c をタイプ→　shutdownします？と聞いてくるので， y　→ エンター　で完全終了．<br>
　　　<span style="color: blue;">注意）　(4)をしないと動作し続けてしまうので，必ず(4)まで実行してください．</span>
   


(4) Ctr-c の後に，以下のコメントが表示されます。　ここで，  y をタイプ。


5秒をすぎると，　消えてしまいます。
その場合は，再度 Ctr-c をタイプしてみてください．
