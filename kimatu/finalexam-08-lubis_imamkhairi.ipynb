1日本語を入力するとウチュウ（宇宙）語に翻訳するコードを作成する問題です．  
　ウチュウ語に翻訳する規則を次のように設定します．   

**規則：** すべてをカタカナに変換し，「助詞」を取り除く

 問題文中の"l"は行番号を表す記号を表します．

l14:「ヨミガナを返す関数にtokenを渡しヨミガナを取得」, l15:「ウチュウ語の結果を返すように書く」，l19:「適切な品詞を書く」を適切なコードに書き換えて日本語（一文）をウチュウ語（一文）に変換する関数nihongo2ucyugo()と単語のヨミガナを返す関数token2ucyugo()を定義し，以下の文をnihongo2ucyugo()を用い連続技で翻訳せよ．  

1. われわれは地球を救うためにエム78星雲からやってきた。
2. 警察は自転車で逃げる泥棒を追いかけた。
3. 私をベクトルにできる者はおるか。

```python
#形態素解析にはjanomeを用います
from janome.tokenizer import Tokenizer

# Tokenizerインスタンスの生成 
tokenizer = Tokenizer()

# 日本語をウチュウにする関数
def nihongo2ucyugo(input_str):
  # 形態素解析の実施
  tokens = tokenizer.tokenize(input_str)
  # 各token の変換結果を" "（半角スペース) でつなげる
  result_str = ""
  for token in tokens:
    result_str += token2ucyugo(token) + " "
  return result_str

# tokenが助詞の場合は空文字列、それ以外はヨミガナを返す関数
def token2ucyugo(input_token):
  if input_token.part_of_speech.split(',')[0] == "助詞":
    return ""
  else:
    return input_token.reading

text = ["われわれは地球を救うためにエム78星雲からやってきた。",
        "警察は自転車で逃げる泥棒を追いかけた。",
        "私をベクトルにできる者はおるか。"]

#動作確認のプリント文挿入
print(nihongo2ucyugo(text[0]))
```

```python
#連続技で実行
for sentence in text:
  print(nihongo2ucyugo(sentence))
```

## 問4
\[評価値\]<br>
映画評価コメントをポジティブな意見かネガティブな意見をまとめたデータに対してロジスティック回帰モデルを適用し，正解のラベル，予測されたラベル，予測確率をタブ区切り形式で出力したファイル result.txt を利用し，その予測結果を次の評価値で評価せよ。<br>
1.正解率，2.適合率，3.再現率および 4.F値

```python
fname_result = 'result.txt' 

def score(fname):
    '''結果ファイルからスコア算出
    結果ファイルを読み込んで、正解率、適合率、再現率、F1スコアを返す
    戻り値：
    正解率,適合率,再現率,F1スコア
    '''
    TP = 0      # True-Positive     予想が+1、正解も+1
    FP = 0      # False-Positive    予想が+1、正解は-1
    FN = 0      # False-Negative    予想が-1、正解は+1
    TN = 0      # True-Negative     予想が-1、正解も-1

    with open(fname) as data_file: #ファイルをopenし，data_fileとする
        for line in data_file:
            cols = line.split('\t')    # "-1  -1  0.7992886809287588"の並びのデータをtabで分割しリスト化
#            print(cols)   #colsの内容を表示

            if len(cols) < 3:
                continue

            if cols[0] == '+1':         # 0番目：正解極性
                if cols[1] == '+1':     # 1番目：予想極性
                    TP += 1
                else:
                    FN += 1
            else:
                if cols[1] == '+1':
                    FP += 1
                else:
                    TN += 1

    accuracy = (TP + TN) / (TP + FN + FP + TN)    # 正解率
    precision = TP / (TP+FP)     # 適合率
    recall = TP/(TP+FN)    # 再現率
    f1 = (2*precision*recall) / (recall + precision)   # F1スコア

    return accuracy, precision, recall, f1

accuracy, precision, recall, f1 = score(fname_result)
print('正解率　\t{}\n適合率　\t{}\n再現率　\t{}\nF1スコア　\t{}'.format( accuracy, precision, recall, f1))
```

**問：**　この結果について，考察してください． 「〇〇のことから，△△といえる」のような根拠から導出される結論の文で答えること。この文型に合わせなくても根拠からの結論が記述されていれば構わない。



**解答：**
正確率は 70％ 超えていることから、このデータは結構正確であることといえる。90％ぐらいの正確率よりあまり正確とも言えない。

\[単語の重要度\]<br>
以下の文書0〜3の4文を用いて，各素性（単語）のTF-IDFを求めよ。TF-IDFをpandasライブラリを利用してデータフレームを用い表示し<span style="color: skyblue; ">（最終的に，そのほかの配列や値は表示する必要はない．）</span>，文書0~3の特徴のある単語( 𝑡𝑓⋅𝑖𝑑𝑓>0.45 とした場合)を答えてください。複数ある場合は，カンマで区切ること。<br>
手法はTF-IDFの式を直接書く方法でもTfidfVectorizerを用いる方法でもどちらを選択しても構わない．

- 0: 'This is the first document.'
- 1: 'This document is the second document.'
- 2: 'And this is the third one.'
- 3: 'Is this the first document?'

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import normalize #データの正規化などが可能な
import numpy as np #数値計算用ライブラリpnumpyをnpと名前をつけて以下で利用
import pandas as pd #データ解析用ライブラリpandasをpdと名前をつけて以下で利用

corpus = [
    'This is the first document.',
    'This document is the second document.',
    'And this is the third one.',
    'Is this the first document?'
]

smooth_idf = True
norm_idf = True

wc = CountVectorizer()
x = wc.fit_transform(corpus)
wcX = np.array(x.toarray())

print('4文コーパス: \n', corpus, '\n') 
print('頻度ベクトル: \n', wcX,'\n')  
```

```python
N = wcX.shape[0]

tf = np.array([wcX[i, :] / np.sum(wcX, axis=1)[i] for i in range(N)])

print('4文コーパス: \n', corpus, '\n') 
print('全文書数： ', N, '\n') 


print('wcX[i, :] \n', [wcX[i, :]for i in range(N)], '\n')
print('words_freq \n', [np.sum(wcX, axis=1)[i]for i in range(N)], '\n')
print('tf \n', tf, '\n')

df = np.count_nonzero(wcX, axis=0) 
print('df: ', df, '\n') 

idf = np.log((1 + N) / (1 + df) )+ 1  if smooth_idf else np.log( N / df )  
print('idf :', idf, '\n')

tfidf = normalize(tf*idf) if norm_idf else tf*idf

df_tfidf = pd.DataFrame(tfidf, columns=wc.get_feature_names())

print(df_tfidf)
```

```python
def getCharacteristicWord():
  result = [[] for i in range(len(df_tfidf["and"]))]

  for column in df_tfidf.columns:
    for i in range(len(df_tfidf[column])):
      if (df_tfidf[column].iloc[i] > 0.45):
        result[i].append(column)

  for i in range(len(result)):
    print("文書",i,result[i])
```
